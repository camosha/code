```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
tickers = pd.read_html(
    'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
tickers
```

```python
tickers = tickers.Symbol.to_list()
```

```python
tickers
```

```python
df = pd.DataFrame()
for ticker in tickers:  
  x = yf.Ticker(ticker).info
  new_frame = pd.DataFrame([x])
  df = df.append(new_frame)

```

```python
engine = sqlalchemy.create_engine('sqlite:///Fundamentals.db')
```

```python
df
```

```python
subdf = df[['symbol', 'dividendYield', 'sector', 'shortRatio', 'forwardPE', 'trailingPE']]
```

```python
subdf
```

```python
subdf.to_sql('Fundamentalstable',engine,index=False)
```

```python
pd.read_sql('Fundamentalstable',engine, index_col=['symbol'])
```
